```python
import numpy as np
import pandas
import scipy.linalg
```

```python
def whitening_transform(X, lambda_):
    C = (1/X.shape[0]) * np.dot(X.T, X)
    s, V = scipy.linalg.eigh(C)
    D = np.diag( 1. / np.sqrt(s + lambda_) )
    W = np.dot( np.dot(V, D), V.T)
    return W

def whiten(X, mu, W):
    return np.dot( X - mu, W)
```

```python
path = "/data1/santiago/BBBC021/deep_profiler/collapsed/cp-segmentation/Wells.csv"
wells = pandas.read_csv(path)
```

```python
metadata = pandas.read_csv("/data1/santiago/BBBC021/metadata/metadata.csv")
```

```python
metadata = metadata.drop_duplicates(['Metadata_Plate', 'Metadata_Well'])
```

```python
metadata
```

```python
merged = pandas.merge(wells, metadata, left_on=['Plate', 'Well'], right_on=['Metadata_Plate', 'Metadata_Well'])
```

```python
merged
```

```python
control = merged[merged['Image_Metadata_Compound'] == 'DMSO']
```

```python
control = control[wells.columns].drop(['Plate', 'Well'], axis=1)
```

```python
X = np.asarray(control)
```

```python
mu = X.mean(axis=0)
```

```python
W = whitening_transform(X - mu, 1e-6)
```

```python
Z = np.asarray(wells.drop(['Plate', 'Well'], axis=1))
```

```python
whitened = whiten(Z, mu, W)
```

```python
data = pandas.DataFrame(whitened)
```

```python
data = pandas.concat([data, platewell], axis=1)
```
