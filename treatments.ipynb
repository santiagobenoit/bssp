```python
# irn
import pandas
import os
```

```python
wells = pandas.read_csv("/data1/santiago/BBBC021/deep_profiler/collapsed/Wells.csv")
metadata = pandas.read_csv("/data1/santiago/BBBC021/metadata/metadata.csv")
moa = pandas.read_csv("/data1/santiago/BBBC021/metadata/BBBC021_v1_moa.csv")
```

```python
moa
```

```python
compounds = metadata.drop_duplicates(['Image_Metadata_Compound', 'Image_Metadata_Concentration', 'Metadata_Plate', 'Metadata_Well'])[['Image_Metadata_Compound', 'Image_Metadata_Concentration', 'Metadata_Plate', 'Metadata_Well']]
```

```python
merged = pandas.merge(compounds, wells, right_on=['Plate', 'Well'], left_on=['Metadata_Plate', 'Metadata_Well'])
```

```python
mean = merged.groupby(['Image_Metadata_Compound', 'Image_Metadata_Concentration']).mean()
```

```python
mean = mean.reset_index()
mean
```

```python
treatments = pandas.merge(mean, moa, left_on=['Image_Metadata_Compound', 'Image_Metadata_Concentration'], right_on=['compound', 'concentration'])
treatments = treatments.drop(['Image_Metadata_Compound', 'Image_Metadata_Concentration'], axis=1)
```

```python
cols = treatments.columns.tolist()
cols = cols[-3:] + cols[:-3]
treatments = treatments[cols]
```

```python
treatments
```

```python
treatments.to_csv("/data1/santiago/BBBC021/deep_profiler/collapsed/Treatments.txt", index=False, sep='\t')
```
